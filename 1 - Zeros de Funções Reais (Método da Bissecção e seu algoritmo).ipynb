# Método de Bissecção
É um método numérico para encontrar raiz de uma função.

Seja y = f(x) uma função contínua em um intervalo [a,b] que contém uma, e só uma, raiz, , da equação f(x) = 0. Este método consiste em dividir o intervalo [a, b], de forma iterativa, ao meio. Para verificar se a raiz está contida na primeira ou na segunda metade do intervalo inicial, é utilizado o teorema de Bolzano. Em seguida, o processo é repetido para aquela metade que contém a raiz de f(x) = 0, ou seja, aquela em que a função, y = f(x), tem valores numéricos com sinais opostos nos seus extremos.
 
## Função de iteração
Considerando que em cada iteração é atualizado o ponto “a” ou “b”, tem-se que a função
de iteração

## Critério de parada
Dada uma precisão ξ, o processo iterativo é finalizado quando se obtém um intervalo cujo
tamanho é menor ou igual a ξ, então qualquer ponto nele contido pode ser tomado como
uma estimativa para a raiz; ou quando for atingido um número máximo de iterações.

## Critério de convergência
Se y = f(x) for contínua em [a, b] e f(a).f(b) < 0, então o método da Bisseção gera uma seqüência que converge para uma raiz de f(x) = 0.

## Estimativa do número de iterações
O método da Bisseção permite que seja estimado, a priori, o número mínimo de iterações
para calcular uma raiz ξ com uma precisão ξ a partir de um intervalo [a, b]. As iterações geram uma seqüência de intervalos encaixados da forma

# Teorema de Bolzano
Se f(a) * f(b) < 0, então existe pelo menos uma raiz em [a.b]



```python
# Importar a biblioteca matemática
import math
```

```python
# Definir um intervalo [a,b] e um erro
a = 2
b = 3
e = 0.01 # definir o erro
```

```python
# Definir a função
def f(x):   # a função se chama f que recebe um parâmetro chamado x
    return x**2 - 5 # ** X ao quadrado - 5 (calcular a raiz quadrada de 5)

# Teste do teorema de Bolzado: ... "Se é uma função contínua num intervalo fechado, 
# e k um número real compreendido entre e, então existe pelo menos um valor real c, 
# pertencente ao intervalo aberto tal que = k ".

# Resumindo: precisamos verificar se o intervalor a b existe mesmo a raiz da função.

if f(a) * f(b) < 0:                 # Se o produto for menor que 0, será executado o código.
    while (math.fabs(b-a)/2 > e):   # Uso do laço para repetição. Condição que seja verdadeiro. O módulo de (b-a)/2 for maior que o erro.
        xi = (a+b)/2                # Quando for menor, já atingiu o valor da raiz aproximada.
        if f(xi) == 0:              # xi é a raiz da função. Se for verdadeiro, sairá do laço com o break
            print('A raiz é: ',{xi})
            break
        else:                       # Caso não for verdade, usa-se o teorema de Bolzado novamente. 
            if f(a) * f(xi) < 0:    # Vamos verificar com base no intervalo inicial dividido ao meio, se a
                b = xi                 # raiz está do lado esquerdo ou do lado direito.
            else:
                a = xi
    print('Valor da raiz é: ',{xi})
else: 
    print('Não existe raiz neste intervalo!') # Caso contário, não existirá raiz.
```
