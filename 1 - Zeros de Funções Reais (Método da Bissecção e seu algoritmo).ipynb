# Mtodo de Bisseco
 um mtodo numrico para encontrar raiz de uma funo.

Seja y = f(x) uma funo contnua em um intervalo [a,b] que contm uma, e s uma, raiz, , da equao f(x) = 0. Este mtodo consiste em dividir o intervalo [a, b], de forma iterativa, ao meio. Para verificar se a raiz est contida na primeira ou na segunda metade do intervalo inicial,  utilizado o teorema de Bolzano. Em seguida, o processo  repetido para aquela metade que contm a raiz de f(x) = 0, ou seja, aquela em que a funo, y = f(x), tem valores numricos com sinais opostos nos seus extremos.
 
## Funo de iterao
Considerando que em cada iterao  atualizado o ponto a ou b, tem-se que a funo
de iterao

## Critrio de parada
Dada uma preciso , o processo iterativo  finalizado quando se obtm um intervalo cujo
tamanho  menor ou igual a , ento qualquer ponto nele contido pode ser tomado como
uma estimativa para a raiz; ou quando for atingido um nmero mximo de iteraes.

## Critrio de convergncia
Se y = f(x) for contnua em [a, b] e f(a).f(b) < 0, ento o mtodo da Bisseo gera uma seqncia que converge para uma raiz de f(x) = 0.

## Estimativa do nmero de iteraes
O mtodo da Bisseo permite que seja estimado, a priori, o nmero mnimo de iteraes
para calcular uma raiz  com uma preciso  a partir de um intervalo [a, b]. As iteraes geram uma seqncia de intervalos encaixados da forma

# Teorema de Bolzano
Se f(a) * f(b) < 0, ento existe pelo menos uma raiz em [a.b]



```python
# Importar a biblioteca matemtica
import math
```

```python
# Definir um intervalo [a,b] e um erro
a = 2
b = 3
e = 0.01 # definir o erro
```

```python
# Definir a funo
def f(x):   # a funo se chama f que recebe um parmetro chamado x
    return x**2 - 5 # ** X ao quadrado - 5 (calcular a raiz quadrada de 5)

# Teste do teorema de Bolzado: ... "Se  uma funo contnua num intervalo fechado, 
# e k um nmero real compreendido entre e, ento existe pelo menos um valor real c, 
# pertencente ao intervalo aberto tal que = k ".

# Resumindo: precisamos verificar se o intervalor a b existe mesmo a raiz da funo.

if f(a) * f(b) < 0:                 # Se o produto for menor que 0, ser executado o cdigo.
    while (math.fabs(b-a)/2 > e):   # Uso do lao para repetio. Condio que seja verdadeiro. O mdulo de (b-a)/2 for maior que o erro.
        xi = (a+b)/2                # Quando for menor, j atingiu o valor da raiz aproximada.
        if f(xi) == 0:              # xi  a raiz da funo. Se for verdadeiro, sair do lao com o break
            print('A raiz : ',{xi})
            break
        else:                       # Caso no for verdade, usa-se o teorema de Bolzado novamente. 
            if f(a) * f(xi) < 0:    # Vamos verificar com base no intervalo inicial dividido ao meio, se a
                b = xi                 # raiz est do lado esquerdo ou do lado direito.
            else:
                a = xi
    print('Valor da raiz : ',{xi})
else: 
    print('No existe raiz neste intervalo!') # Caso contrio, no existir raiz.
```
